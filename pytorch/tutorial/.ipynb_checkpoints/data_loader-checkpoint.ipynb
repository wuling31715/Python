# Data Loader

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms, utils
from PIL import Image
```

```python
def img_load(path):
    img = Image.open(path).convert('RGB')
    return img
```

```python
def img_show(img):
    plt.imshow(img)
    plt.show()
```

```python
img_list = list()
for i in range(10000):
    img_path = 'halftone32/x_test/%s.png' % str(i)
    img = img_load(img_path)
    img_list.append(img)
```

```python
y_train = np.load('halftone32/y_train.npy') 
y_train = y_train.tolist()
```

```python
y_test = np.load('halftone32/y_test.npy') 
y_test = y_test.tolist()
```

```python
transform = transforms.Compose(
    [transforms.ToTensor()]
)
```

```python
class ListData(Dataset):
    
    def __init__(self, X_list, y_list, transform):
        self.X_list = X_list
        self.y_list = y_list
        self.transform = transform
        if len(self.X_list) == len(self.y_list):
            print('len(X_list) == len(y_list)')
        else:
            print('len(X_list) != len(y_list)')
        print('transform: %s' % self.transform)
            
    def __getitem__(self, index):
        if self.transform != None:
            X = self.transform(self.X_list[index])
            y = self.y_list[index]
        return X, y
    
    def __len__(self):
        return len(self.y_list)
```

```python
data_test = ListData(img_list, y_test, transform)
```

```python
data_loader_test = torch.utils.data.DataLoader(
    dataset=data_test,
    batch_size=64,
    shuffle=True,
)
```

```python
for data in data_loader_test:
    X_test, y_test = data
    print(X_test, y_test)
    break
```
