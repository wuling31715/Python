```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
import os
from PIL import Image
```

```python
files = ['lenna.png', 'lenna.png', 'lenna.png', 'lenna.png']
```

```python
result = Image.new("RGB", (800, 800))
for index, file in enumerate(files):
    path = os.path.expanduser(file)
    img = Image.open(path)
    img.thumbnail((400, 400), Image.ANTIALIAS)
    x = index // 2 * 400
    y = index % 2 * 400
    w, h = img.size
    print(x, y, x + w, y + h)
    result.paste(img, (x, y, x + w, y + h))
imshow(result)
result.save(os.path.expanduser('square.png'))
```

```python
result = Image.new("RGB", (800, 200))
for index, file in enumerate(files):
    path = os.path.expanduser(file)
    img = Image.open(path)
    img.thumbnail((200, 200), Image.ANTIALIAS)
    x = index * 200
    w, h = img.size
    print(x, 0, x + w, h)
    result.paste(img, (x, 0, x + w, h))
imshow(result)
result.save(os.path.expanduser('horizontal.png'))
```
